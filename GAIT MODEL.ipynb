```python
import zipfile
import os

# Specify the path to the downloaded zip file
zip_file_path = '/content/human-gait-phase-dataset.zip'

# Specify the directory to extract the contents
extracted_path = '/content/'

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

# List the files in the extracted directory
os.listdir(extracted_path)
```

```python
df1 = pd.read_csv('/content/data/GP1_0.6_marker.csv')
df2 = pd.read_csv('/content/data/GP1_0.7_marker.csv')
```

```python
df1
```

```python
df2
```

```python
data1 = pd.DataFrame({'secim':np.zeros(12000)})
data2 = pd.DataFrame({'secim':np.ones(12000)})

data2 = pd.concat([df2,data2],axis=1)
data1 = pd.concat([df1,data1],axis=1)

data = data1.append(data2,ignore_index=True)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.histplot(data['secim'], kde=True)
plt.title('Distribution of Target Variable (secim)')
plt.show()
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True)
plt.title('Correlation Matrix')
plt.plot(figsize=(20,10))
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(data.columns[:-1]):
    plt.subplot(5, 5, i+1)
    sns.boxplot(x='secim', y=column, data=data)
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(data.columns[:-1]):
    plt.subplot(5, 5, i+1)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
x = data.drop(["secim"],axis=1)
y = data.secim.values
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

SVM

```python
from sklearn.svm import SVC

svm = SVC(random_state = 1)
svm.fit(x_train,y_train)
print("acc of svm is :",svm.score(x_test,y_test))
```

Decision Tree Classifier.

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print('Accuracy of dec tree in test data is:',dt.score(x_test,y_test))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100,random_state=1)
rf.fit(x_train,y_train)
print('Random Forest accuracy on test data is : ',rf.score(x_test,y_test))
```
